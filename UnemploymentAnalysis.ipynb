Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv("Unemployment in India.csv")
```

```python
df
```

```python
df.describe()
```

```python
df.info()
```

Checking for null Values and Dropping them

```python
null_values=df.isnull().sum()
print(null_values)
```

```python
cleaned_rows=df.dropna(inplace= True)
cleaned_rows
```

```python
df[' Date'] =pd.to_datetime(df[' Date'])
```

```python
df[' Estimated Unemployment Rate (%)'].mean()
```

```python
df
```

```python
group_by_region_df= df.groupby(by='Region').mean().sort_values(by=' Estimated Unemployment Rate (%)',ascending=False)
group_by_region_df
group_by_region_df.sort_values(by=' Estimated Unemployment Rate (%)',ascending=False)
```

Analyzing the Unemployment rate of a particular Region

```python
unemp_state_df=df[df.Region=='Delhi']
unemp_state_df.head()
```

```python
unemp_state_df[' Estimated Unemployment Rate (%)'].mean()
```

```python
delhi_2020=unemp_state_df.loc[unemp_state_df[' Date'].dt.year == 2020]
delhi_2020[' Estimated Unemployment Rate (%)'].mean()
```

```python
delhi_2020.sort_values(by=' Estimated Unemployment Rate (%)',ascending=False).head(1)
```

```python
df['year'] = pd.DatetimeIndex(df[' Date']).year
df['month'] = pd.DatetimeIndex(df[' Date']).month
```

```python
years=list(set(list(df[' Date'].dt.year)))
gbyyear_df=df.groupby(by='year')
mean_uemployment_rate_overall =list(gbyyear_df[' Estimated Unemployment Rate (%)'].mean())
mean_uemployment_rate_overall
```

Visualization of the Dataset

```python
plt.figure(figsize=(10,7))
sns.heatmap(df.corr(), cmap='viridis')
plt.show()
```

```python
df.head()
```

```python
sns.set_style("darkgrid")
plt.plot(years, mean_uemployment_rate_overall, 's-b')
plt.xlabel('Year')
plt.ylabel('mean_uemployment_rate_overall(%)')

plt.title("Unemployment rate by Year")

```

Analyzing Unemployment rate based on Months

```python
gbm=df.groupby(by='month', as_index=False).mean()
gbm
```

```python
sns.lmplot(x='month',y=' Estimated Unemployment Rate (%)',height=5,data=gbm);
```

```python
graph = sns.PairGrid(group_by_region_df.sort_values(" Estimated Unemployment Rate (%)", ascending=False),
                 x_vars=group_by_region_df.columns[1:4], y_vars=["Region"],
                 height=15)
graph.map(sns.stripplot, size=12, orient="h", jitter=False,
      palette="flare_r", linewidth=1, edgecolor="w")
```
